#### Task 1
Create a function which for a given list of integer numbers finds the longest chain of consecutive values of an arithmetic sequence with difference equal to 5. For example, for the list
$$
[1,2,7,12,6,3,1,6,11,16,7]
$$
the answer should be $4$, as the chain we are looking for is $1,6,11,16$. Note that there is also a shorter chain $2,7,12$, so you function should take into account that there may be a shorter sequence satisfying the conditions of the task somewhere earlier in the list.

````python
def find_consecutives(numbers: list) -> int:
    if not numbers:
        return 0
    
    count = 0
    max_count = 0
    previous = numbers[0]

    for nb in numbers[1:]:
        if previous + 5 == nb:
            count += 1
            max_count = max(max_count, count)
        else:
            count = 0
        
        previous = nb
    
    return max_count + 1
````

````python
numbers = [1, 2, 12, 6, 3, 15, 6, 11, 16, 7]
print(find_consecutives(numbers))
````

#### Task 2
In this task we assume the standard latin alphabet consisting of 26 letters. We introduce the following method of encryption:
- a letter from the first half of the alphabet, that is, from a to m, is shifted by 13,
- a letter from the second half of the alphabet, that is, from n to z, is mirror-reflected into the first half of the alphabet.

For example, $a\mapsto n$, $b\mapsto o,\ldots,m\mapsto z$, while $n\mapsto m$ (since n is the first letter in the second half and m is the last letter in the first half), $o\mapsto l,\ldots, z\mapsto a$. Create a function which for a given string, encodes the string according to this algorithm.

````python
alphabet = 'abcdefghijklmnopqrstuvwxyz'

def encryption(text: str) -> str:
    new_str = []
    for letter in text:
        if letter.lower() not in alphabet:
            new_str.append(letter)
            continue

        index = alphabet.index(letter.lower())
        if index < 13:
            new_str.append(alphabet[index + 13])
        else:
            new_str.append(alphabet[25 - index])
    
    return ''.join(new_str)
````

````python
text = 'ALGORITHM'

print(encryption(text))
````

````python
def create_adjacency_list(edges: list) -> dict:
    adj = {}
    for u, v in edges:
        if u not in adj:
            adj[u] = set()
        if v not in adj:
            adj[v] = set()
        
        adj[u].add(v)
        adj[v].add(u)
    
    return adj
````

Implement a class which represents complex numbers. Let us recall that a complex number is a number of the from $z=a+ib$, where $a=Re(z)$ is called the real part of $z$, and $b=Im(z)$ is called the imaginary part of $z$. Each complex number is uniquely determined by its real and imaginary parts and thus can be identified with a pair of real numbers $z\approx(a,b)$. Your class must contain:
- the method __init__ which upon creating an instance of the class assigns the values of $Re(z)$ and $Im(z)$,
- a method which prints the form of the number represented by an instance (i.e. it should print a string of the form $a+ib$)
- a method which adds two complex numbers represented by two instances of the class (let us recall that for $z_1=a+ib$ and $z_2=x+iy$ the sum is defined as $z_1+z_2=(a+x)+i(b+y)$.


````python
class ComplexNumber:
    def __init__(self, real, imag):
        """Inicjalizacja części rzeczywistej i urojonej."""
        self.real = real
        self.imag = imag

    def display(self):
        """Metoda do wypisywania liczby w formacie a + ib."""
        # Dodajemy warunek dla znaku, żeby nie wyszło np. 2 + i-3
        sign = "+" if self.imag >= 0 else "-"
        print(f"{self.real} {sign} i{abs(self.imag)}")

    def add(self, other):
        """Dodaje dwie liczby zespolone i zwraca nowy obiekt."""
        new_real = self.real + other.real
        new_imag = self.imag + other.imag
        return ComplexNumber(new_real, new_imag)

# --- TESTY ---
z1 = ComplexNumber(3, 5)   # 3 + i5
z2 = ComplexNumber(1, -2)  # 1 - i2

print("Liczba 1:")
z1.display()

print("Liczba 2:")
z2.display()

suma = z1.add(z2)
print("Suma:")
suma.display() # Powinno być: 4 + i3
````
