1. The Great Power of a natural number n (denoted by WS(n)) is the number of zeroes the number n! ends with. Write a function which for a given number n calculates WS(n). To facilitate the task, we have prepared some tests:<br>
$GP(1000)=249$ <br>
$GP(4)=0$,<br>
$GP(60)=14$.


```python
import math

def GP(n):
    count = 0
    while n >= 5:
        n //= 5
        count += n
    return count


# --- DEBUGING ---
print(GP(1000))
print(GP(4))
print(GP(60))
```

2. A numbersaur is a number which is written backwards. For example, the numbersaur 7431 in fact equals 1347. Your task is to prepare a function, which adds two numbersaurs and prints the result of this addition in two ways (for example as a tuple) -- the number itself and its numbersaur. Below you can find some examples to help you understand the task.<br>
$24+_{\text{rawr}}1=(43,34)$ as the numbersaurs 24 and 1 correspond to 42 and 1, and the sum of these numbers equals 43. The numbersaur of this number equals 34.<br>
$4358+_{\text{rawr}}754=(8991,1998)$<br>
$305+_{\text{rawr}}794=(1000,1)$ since the numbersaur of the number 1000 equals 0001, which will be interpreted by a computer as 1.

```python
def reverse_number(n):
    r = 0
    while n > 0:
        r *= 10
        r += n % 10
        n //= 10
    return r


def add_numbersaur(nb1: int, nb2:int) -> tuple:
    '''
    Adds two numbersaurs of given numbers and returns calculated value and its numbersaur in a form of a tuple.
    '''
    result = reverse_number(nb1) + reverse_number(nb2)
    return (result, reverse_number(result))

# --- DEBUGING ---
print(add_numbersaur(24, 1))
print(add_numbersaur(4358, 754))
print(add_numbersaur(305,794))
```

3. Write a function which for a given string returns every second character from the second half of the string (you can assume that we will pass to this function only strings of length divisible by 4).

```python
def manipulate_str(txt: str) -> str:
    return ''.join([letter for i, letter in enumerate(txt[len(txt)//2:]) if i % 2 == 0])
```

```python
print(manipulate_str('abcdefghijklm'))
```

4. A circle or a disk in the plane can be stored in Python as a single tuple of the form  (x,y,r), where $x,y\in\mathbb{R}$ are the first and the second coordinate of the center of the circle\disk, respectively, and  r>0 represents its radius. Write a function which for a pair of circles $O_1$ and $O_2$ finds the positional relation between them:
- the open disk determined by $O_1$ is contained in $O_2$ or vice versa, and the boundaries of these disks (the circles) do not intersect -- in this case the function should print 'Internal';<br>
- the open disk determined by $O_1$ is contained in $O_2$ or vice versa, but the circles intersect in exactly one point, i.e., $O_1\cap O_2={(x_0,y_0)}$ -- in this case the function should print 'Edge';<br>
- in other cases the function should print 'Other'.<br>

    You can assume that all inputs consist of only integer values.

```python
def disk_relation(O1: tuple, O2: tuple) -> str:
    dist = ((O1[0] - O2[0]) ** 2 + (O1[1] - O2[1]) ** 2) ** (1/2)
    radius = abs(O1[2] - O2[2])

    if dist < radius:
        return 'Internal'
    elif dist == radius:
        return 'Edge'
    else:
        return 'Other'
```

```python
print(disk_relation((1, 1, 36), (0, 0, 2)))
```

5. A finite sequence of numbers $[a_0,\ldots,a_n]$ is called a hill, if there exists an index $0\leq j\leq n$ such that:
- the sequence $[a_0,\ldots,a_j]$ is nondecreasing,
- the sequence $[a_j,\ldots,a_n]$ is nonincreasing.<br>

    Write a function which for a given argument $X$ being a table representing the considered sequence returns the highest index j, for which we can divide $X$ in a way which shows that X is a hill. In the case when such a division is not possible, the function should print 'Impossibru'.

```python
def is_hill(X):
    n = len(X)

    for j in range(n - 1, -1, -1):
        
        left_side_ok = True
        for i in range(j):
            if X[i] > X[i+1]:
                left_side_ok = False
                print(f'Dla {i} lewej strony: {left_side_ok}')
                break
        
        if not left_side_ok:
            continue
            
        right_side_ok = True
        for i in range(j, n - 1):
            if X[i] < X[i+1]:
                right_side_ok = False
                break
        
        if left_side_ok and right_side_ok:
            return j

    return "Impossibru"

# --- TESTS ---
print(is_hill([1, 2, 4, 4, 2, 1])) # Should be 3
# print(is_hill([1, 2, 1, 2]))       # Should be Impossibru
# print(is_hill([5, 5, 5]))          # Should be 2
```
