# Exercise 1. Classy polynomials

Implement a class ```Polynomial``` to represent and manipulate with polynomials of the form:

$$ f(x) := a_0 + a_1x + \ldots + a_{n-1}x^{n-1} + a_nx^n $$

Objects of the class ```Polynomial``` should store the values of the coefficients $a_0,\dots, a_n$.

The class should allow to:

- evaluate the polynomial at a given point $x$,
- differentiate -- a method which returns another object of the class Polynomial representing the derivative of the original polynomial,
- subtract and add two polynomials,
- check whether two polynomials are equal.

````python
class Polynomial:
    def __init__(self, coefficients: list):
        '''
        Docstring for __init__
        
        :param self: Generate the polynomial of the form f(x) = a_0 + a_1x + ... +a_nx^n
        :param coefficients: It must be ordered from left to right, as in the formula.
        :type coefficients: list
        '''
        self.coefficients = coefficients


    def evaluate(self, x) -> int:
        '''
        Return the value for x of a given polynomial.
        
        :param x: The argument for the function f(x)
        '''
        return sum([coeff * x**i for i, coeff in enumerate(self.coefficients)])


    def differentiate(self) -> object:
        '''
        Return the derivative of a current polynomiast as another object.
        '''
        return Polynomial([self.coefficients[i] * i for i in range(1, len(self.coefficients))])
    

    def add(self, polynomial: object):
        '''
        Adds to the current polynomial a different one.
        
        :param self: Current polynomial.
        :param polynomial: Another polynomial.
        :type polynomial: Object
        '''
        c1 = self.coefficients
        c2 = polynomial.coefficients
        max_len = max(len(c1), len(c2))

        self.coefficients = [
            (c1[i] if i < len(c1) else 0) + (c2[i] if i < len(c2) else 0)
            for i in range(max_len)
        ]


    def substract(self, polynomial: object):
        '''
        Substract the current polynomial with a different one.
        
        :param self: Current polynomial.
        :param polynomial: Another polynomial.
        :type polynomial: Object
        '''
        c1 = self.coefficients
        c2 = polynomial.coefficients
        max_len = max(len(c1), len(c2))

        self.coefficients = [
            (c1[i] if i < len(c1) else 0) - (c2[i] if i < len(c2) else 0)
            for i in range(max_len)
        ]

    
    def check_equality(self, polynomial: object) -> bool:
        '''
        Checks whether the two polynomials are equal.
        
        :param self: Current polynomial
        :param polynomial: Given polynomial
        :type polynomial: object
        :return: Returns True or False
        :rtype: bool
        '''
        return self.coefficients == polynomial.coefficients
````

````python
funkcja = Polynomial([3, 2, 5])             # f(x) = 1 + x + x^2
pochodna = funkcja.differentiate()          # f'(x) = 1 + 2x
print(f"Coefficients of f(x): {funkcja.coefficients}")
print(f"Coefficients of f'(x): {pochodna.coefficients}")

funkcja.substract(pochodna)                       # f(x) = 2 + 3x + x^2
print(f"Coefficients of f(x) + f'(x): {funkcja.coefficients}")

funkcja1 = Polynomial([1, -8, 5, 5])
print(funkcja.check_equality(funkcja1))
````

# Exercise 2. He deals the cards as a meditation

Implement a class ```Card```, which represents a single card from the standard deck. Then, implement a class ```Deck```, which stores the full 52-card deck.The class ```Deck``` should contain a method ```draw```, which draws the first card from the top of the deck and a method ```shuffle```, which -- you wouldn't belive -- shuffles the deck (use the method ```shuffle``` from the library ```numpy```).

````python
class Card:
    def __init__(self, rank: int, suit: str):
        '''
        Creates a card with given spiecification.
        
        :param rank: Value of the card. Type from 1 - 13, where 1 is As.
        :type rank: int
        :param suit: Available: club - ♣, diamond - ♦, heart - ♥, spade - ♠
        :type suit: str
        '''
        if rank < 1 or 13 < rank:
            raise Exception("Rank must be between 1 and 13")
        if suit not in ['♣', '♦', '♥', '♠']:
            raise Exception("Suit must fit in the list: ['♣', '♦', '♥', '♠']")
        
        self.rank = rank
        self.suit = suit
        self.names = ["Ace", "Two", "Three", "Four", "Five", "Six", "Seven", 
                      "Eight", "Nine", "Ten", "Jack", "Queen", "King"]
        
    def __str__(self):
        return f"{self.names[self.rank - 1]} {self.suit}"
    
    def __repr__(self):
        return self.__str__()
````

````python
import numpy as np

class Deck:
    def __init__(self):
        '''
        Generates a full deck of 52 cards with 4 suits.
        '''
        suits = ['♣', '♦', '♥', '♠']
        self.deck = [Card(rank, suit) for suit in suits for rank in range(1, 14)]
    
    def draw(self) -> Card:
        '''
        Returns the card from the top of the deck. It removes this card from the deck.
        
        :return: Top card from the deck.
        :rtype: Card
        '''
        return self.deck.pop()
    
    def shuffle(self):
        '''
        Shuffles the deck.
        '''
        np.random.shuffle(self.deck)
````

````python
nasz_deck = Deck()
print(nasz_deck.deck)
nasz_deck.shuffle()
print(nasz_deck.deck)
````

# Writing code during classes

Brief introduction into Classes in Python.

````python
class BankClient:
    
    def __init__(self, name, surname, balance):
        self.name = name
        self.surname = surname
        self.balance = balance
    
    def withdraw(self, amount):
        self.balance -= amount
    
    def deposit(self, amount):
        self.balance += amount
````

````python
new_client = BankClient('Mateusz', 'Konieczny', 10000) # <- creatin an INSTANCE of a class
print(new_client.balance)
````
