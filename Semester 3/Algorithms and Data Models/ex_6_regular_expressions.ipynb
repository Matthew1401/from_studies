#### Exercise 2 -- coronavirus in numbers vol 2
Extract all numbers from the text. This time make sure that each number containing a whitespace is counted as a single number.

```python
numbers_from_text_2 = []

for i in analysed_text.split():
    if i.isdigit():
        numbers_from_text_2.append(int(i))

print(numbers_from_text_2)
```

#### Exercise 3 -- voivodeships vs coronavirus

#### Exercise 4 -- coronavirus: grammatical cases
Find all the grammatical cases of the words 'koronawirus' and 'przypadek' in the text. (Erasmus students -- sorry for that :) )

```python
all_gramatical_cases = set()

for i in analysed_text.split():
    if ('przypad' in i or 'koronawirus' in i) and i.isalpha():
        all_gramatical_cases.add(i)

print(all_gramatical_cases)
```

#### Exercise 5 -- coronavirus: dates
Find all dates in the format day-month-year.

```python
months =['stycznia', 'lutego', 'marca', 'kwietnia', 'maja', 'czerwca', 
         'lipca', 'sierpnia', 'września', 'października', 'listopada', 'grudnia']
dates = []

# splited text
st = analysed_text.split()

for i in range(len(st)):
    if st[i] in months and st[i - 1].isdigit() and st[i + 1].isdigit():
        dates.append(' '.join([st[i - 1], st[i], st[i + 1], st[i + 2]]))

print(dates)
```

#### Exercise 6 -- sentences
Count all sentences in the text.
