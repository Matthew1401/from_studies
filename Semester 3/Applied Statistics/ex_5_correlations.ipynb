```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn import datasets
```

```python
wine = sk.datasets.load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df
```

```python
print("Mean:", round(df['od280/od315_of_diluted_wines'].mean(), 2))
print("Median:", df['od280/od315_of_diluted_wines'].median())
print("Minimum:", df['nonflavanoid_phenols'].min())
print("Standard Deviation:", df['nonflavanoid_phenols'].std())
print("First Quartile (Q1):", df['nonflavanoid_phenols'].quantile(0.25))
print("Third Quartile (Q3):", df['nonflavanoid_phenols'].quantile(0.75))
```

```python
plt.boxplot(df['od280/od315_of_diluted_wines'])
plt.title("Boxplot ")
plt.ylabel("od280/od315_of_diluted_wines")
plt.show()
```

```python
# Generate and display the correlation matrix
correlation_matrix = df.corr()
print("Correlation matrix:\n", correlation_matrix)

# Plot the correlation matrix as a heatmap for better visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, fmt='.2f')
plt.title("Correlation Matrix of Wine Dataset")
plt.show()

# Select two variables for the scatter plot
var1 = 'nonflavanoid_phenols'  # Replace with the actual column name, e.g., 'alcohol'
var2 = 'alcalinity_of_ash'  # Replace with another column name, e.g., 'malic_acid'

# Create a scatter plot for the two selected variables
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=var1, y=var2)
plt.title(f"Scatter Plot of {var1} vs {var2}")
plt.xlabel(var1)
plt.ylabel(var2)
plt.show()
```
