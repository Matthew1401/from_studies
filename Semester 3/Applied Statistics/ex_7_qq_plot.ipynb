```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Generate a sample of size M from real data
sampleR1 = np.array([102, 105, 113, 120, 125, 127, 134, 135, 139, 141, 144, 145, 149, 150, 160]) # Size M=15
sampleR1 = np.sort(sampleR1)

sampleR2 = np.array([
    120, 115, 94, 118, 111, 102, 102, 131, 104, 107,
    115, 139, 115, 113, 114, 105, 115, 134, 109, 109,
    93, 118, 109, 106, 125
]) # Size M=25
sampleR2 = np.sort(sampleR2)
```

```python
# Generate a sample of size N from a normal distribution
# np.random.seed(52)  # Set seed for reproducibility

# sampleN = np.random.normal(loc=0, scale=1, size=20)
# sampleN=np.sort(sampleN)
# print(sampleN)
mean_value = np.mean(sampleR2)
variance_value = np.var(sampleR2, ddof=1)   # sample  variance

print("Sample mean:", mean_value)
print("Sample variance:", variance_value)
```

```python
# Create a QQ-plot
stats.probplot(sampleR2, dist="norm", plot=plt)
plt.title("Q-Q Plot: Normality Check")  # Add a title to the plot
plt.grid(True)  # Enable grid for better readability
plt.show()
```

```python
#  theoretical quantiles and sample quantiles
theoretical_quantiles, ordered_values = stats.probplot(sampleR2, dist="norm")[0]
print("Theoretical quantiles:")
print(theoretical_quantiles)
sample_quantiles=np.sort(sampleR2)
# Compute the correlation coefficient
correlation_coefficient = np.corrcoef(theoretical_quantiles, sample_quantiles)[0, 1]
# Print the correlation coefficient
print("\nCorrelation coefficient:", f"{correlation_coefficient:.4f}")
```
